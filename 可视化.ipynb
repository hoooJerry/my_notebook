## 高级进阶--随机漫步

````python
import matplotlib.pyplot as plt
%matplotlib inline
from random import choice
class RandomWalk():
    """生成随机漫步数据的类"""
    def __init__(self,num_points=5000):
        """初始化随机漫步的属性"""
        self.num_points = num_points
        self.x = [0]
        self.y = [0]
    def fill_walk(self):
        """计算随机漫步包含的所有点"""
    #不断漫步，知道达到指定长度
        while len(self.x) < self.num_points:
            x_direct = choice([1,-1])
            x_distance = choice([1,2,3,4])
            x_step = x_direct*x_distance
            y_direct = choice([1, -1])
            y_distance = choice([1, 2, 3, 4])
            y_step = y_direct * y_distance
        #拒绝原地踏步
            if x_step == 0 and y_step ==0:
                continue
        #计算下一个点的x和y值
            next_x = self.x[-1] +x_step
            next_y = self.y[-1] +y_step
            self.x.append(next_x)
            self.y.append(next_y)

rw = RandomWalk()
rw.fill_walk()
plt.scatter(rw.x,rw.y,s=6)
plt.show()
````

## 高级进阶2--模拟掷骰子
### 使用pygal

````python
import pygal
from random import randint
from IPython.display import SVG, display
class Die():
    def __init__(self,num_side = 6):
        self.num_sides = num_side
    def roll(self):
        """返回一个位于1和面数之间的随机值"""
        return randint(1,self.num_sides)
die = Die()
results = []
for roll in range(100):
    results.append(die.roll())
frequencies =[]
for value in range(1,die.num_sides+1):
    frequency = results.count(value)
    frequencies.append(frequency)
#对结果可视化
hist = pygal.Bar()
hist.title = "result of rolling a die"
hist.x_labels = ['1','2','3','4','5','6']
hist.x_title = "result"
hist.add('D6',frequencies)
display(SVG(hist.render(disable_xml_declaration=True)))
````

### 掷两个骰子

````python
import pygal
from IPython.display import SVG,display
class Die():
    def __init__(self,num_side = 6):
        self.num_sides = num_side
    def roll(self):
        """返回一个位于1和面数之间的随机值"""
        return randint(1,self.num_sides)
die1 = Die()
die2 = Die()
results1 = []
results2 = []
for roll in range(100):
    results1.append(die1.roll())
    results2.append(die2.roll())
frequencies =[]
frequencies2 =[]
for value in range(1,die1.num_sides+1):
    frequency = results1.count(value)
    frequency2 = results2.count(value)
    frequencies.append(frequency)
    frequencies2.append(frequency2)
#对结果可视化
hist = pygal.Bar()
hist.title = "result of rolling a die"
hist.x_labels = ['1','2','3','4','5','6']
hist.x_title = "result"
hist.add('D6a',frequencies)
hist.add('D6b',frequencies2)
display(SVG(hist.render(disable_xml_declaration=True)))
````
